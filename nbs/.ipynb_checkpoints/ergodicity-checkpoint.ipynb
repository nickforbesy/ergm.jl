```julia
using ergm.sampler
using ergm.spaces
using ergm.stats
using ergm.models
using ergm.inference
using ergm.optim
using LinearAlgebra
using Random

n = 30
stats = DeltaStats(
    function(G)
        n = size(G.adjacency, 1)
        [
            sum(G.adjacency) / (n * (n - 1)),
            tr(G.adjacency ^ 3) / (n * (n - 1) * (n - 2))
        ]
    end,
    function(G, s, u)
        n = size(G.adjacency, 1)
        i, x = u
        s + [
            (x - G[i]) / (n * (n - 1)),
            3 * (x - G[i]) * dot(G.adjacency[i[2], :], G.adjacency[:, i[1]]) / (n * (n - 1) * (n - 2))
        ]
    end
)
θ0 = [0.0, 1e4]
m = ExponentialFamily(stats, θ0)
G0 = DiGraph(zeros(Bool, (n, n)))
s = ParallelGibbsSampler(
    G0,
    m,
    10,
    3,
    Threads.nthreads()
);
```

```julia
using Statistics

θ_Ts = map(p -> sign(p) * 10 ^ abs(p), -3:0.5:3)
ms = []
σs = []

for θ_T ∈ θ_Ts
    update_params(m, [0.0, θ_T])
    s = ParallelGibbsSampler(
        G0,
        m,
        10,
        3,
        Threads.nthreads()
    )  
    _, ss = sample(s, 10)
    push!(ms, mean(ss[2, :]))
    push!(σs, std(ss[2, :]))
end
```

```julia
using Plots

plot(-3:0.5:3, ms, yerror=σs)
```
