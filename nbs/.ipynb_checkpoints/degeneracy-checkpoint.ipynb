Define directed edge-triangle model on 30 nodes:

```julia
using ergm.sampler
using ergm.spaces
using ergm.stats
using ergm.models
using LinearAlgebra
using Random

n = 30
stats = DeltaStats(
    function(G)
        p = sum(G.adjacency) / (n * (n - 1))
        [
            p
            tr(G.adjacency ^ 3) / (n * (n - 1) * (n - 2)) - p^3
        ]
    end,
    function(G, s, u)
        i, x = u
        dp = (x - G[i]) / (n * (n - 1))
        dt = 3 * (x - G[i]) * dot(G.adjacency[i[2], :], G.adjacency[:, i[1]]) / (n * (n - 1) * (n - 2))
        s + [
            dp,
            dt + (s[1] + dt) ^ 3 - s[1] ^ 3
        ]
    end
)
θ0 = [0.0, 1e4]
m = ExponentialFamily(stats, θ0)
G0 = DiGraph(zeros(Bool, (n, n)))
```

```julia
using HypothesisTests

function sample_ER(n, p)
    DiGraph(rand(Float64, (n, n)) .< p)
end

function dev_from_ER(Gs)
    stats = SimpleStats(
        function(G)
            [
                sum(G.adjacency) / (n * (n - 1)),
                tr(G.adjacency ^ 3) / (n * (n - 1) * (n - 2))
            ]
        end
    )
    ss = hcat([get_stats(stats, G) for G ∈ Gs]...)
    k = size(ss, 1)
    ms = mean(ss, dims=2)
    p = ms[1]
    ss_ER = hcat([get_stats(stats, sample_ER(n, p)) for _ ∈ 1:k]...)
    ms_ER = mean(ss_ER, dims=2)
    (ms[2] - ms_ER[2]) / ms_ER[2]
end
```

```julia
using Statistics
using ProgressMeter

θs = [[sign(e1) * 10^abs(e1), sign(e2) * 10^abs(e2)] for e1 ∈ -4:0.5:4 for e2 ∈ -4:0.5:4]
ys = []

@showprogress 1 for θ ∈ θs
    update_params(m, θ)
    s = ParallelGibbsSampler(
        G0,
        m,
        10,
        3,
        Threads.nthreads()
    )
    Gs, _ = sample(s, 1000)
    push!(ys, dev_from_ER(Gs))
end
```

```julia
ix = findall(θ -> θ[1] > -100, θs)
surface(map(θ -> θ[1], θs[ix]), map(θ -> θ[2], θs[ix]), ys[ix])
```

```julia
using Plots
p = plot(xs, 100 * ys)
xlabel!("θ_T")
ylabel!("triangle density % diff. from E-R")
png(p, "fig.png")
```
